```python
from qiskit import QuantumCircuit
from qiskit import QuantumRegister, ClassicalRegister
from qiskit.quantum_info import Statevector
```

```python
cr= ClassicalRegister(5,"bell_basis_measurement")
alice= QuantumRegister(1, 'alice__qubit')
bob= QuantumRegister(1, 'bob__qubit')
eve= QuantumRegister(1, 'eve__qubit')
charlie= QuantumRegister(1, 'charlie__qubit')
dave= QuantumRegister(1, 'dave__qubit')
aq1= QuantumRegister(4, 'ancilla_qubit_alice')
aq2= QuantumRegister(4,'ancilla_qubit_bob')
aq3= QuantumRegister(4,'ancilla_qubit_eve')
aq4= QuantumRegister(4,'ancilla_qubit_charlie')
aq5= QuantumRegister(4,'ancilla_qubit_dave')
aq6= QuantumRegister(5,'ancilla_reserve')

repetition5 = QuantumCircuit(alice,aq1,bob,aq2,eve,aq3,charlie,aq4,dave,aq5,aq6,cr)

repetition5.h(alice)
repetition5.cx(alice,bob)
repetition5.cx(alice,eve)
repetition5.cx(alice,charlie)
repetition5.cx(alice,dave)

repetition5.barrier()
repetition5.cx(alice, aq1)
repetition5.cx(bob, aq2)
repetition5.cx(eve, aq3)
repetition5.cx(charlie, aq4)
repetition5.cx(dave, aq5)

repetition5.barrier()

repetition5.x([2,5,9,15,21])
repetition5.barrier()

repetition5.cx(alice, aq1)
repetition5.cx(bob, aq2)
repetition5.cx(eve, aq3)
repetition5.cx(charlie, aq4)
repetition5.cx(dave, aq5)

repetition5.mcx([1,2,3],0)
repetition5.mcx([1,2,4],0)
repetition5.mcx([1,3,4],0)
repetition5.mcx([2,3,4],0)
repetition5.mcx([1,2,3,4],0)
repetition5.mcx([6,7,8],5)
repetition5.mcx([6,7,9],5)
repetition5.mcx([6,8,9],5)
repetition5.mcx([7,8,9],5)
repetition5.mcx([6,7,8,9],5)
repetition5.mcx([11,12,13],10)
repetition5.mcx([11,12,14],10)
repetition5.mcx([11,13,14],10)
repetition5.mcx([12,13,14],10)
repetition5.mcx([11,12,13,14],10)
repetition5.mcx([16,17,18],15)
repetition5.mcx([16,17,19],15)
repetition5.mcx([16,18,19],15)
repetition5.mcx([17,18,19],15)
repetition5.mcx([16,17,18,19],15)
repetition5.mcx([21,22,23],20)
repetition5.mcx([21,22,24],20)
repetition5.mcx([21,23,24],20)
repetition5.mcx([22,23,24],20)
repetition5.mcx([21,22,23,24],20)
repetition5.barrier()

repetition5.cx(alice,bob)
repetition5.cx(alice,eve)
repetition5.cx(alice,charlie)
repetition5.cx(alice,dave)
repetition5.mcx([bob, eve, charlie], aq6[0])
repetition5.mcx([bob, eve, dave], aq6[0])
repetition5.mcx([bob, dave, charlie], aq6[0])
repetition5.mcx([dave, eve, charlie], aq6[0])
repetition5.mcx([bob, eve, charlie, dave], aq6[0])
repetition5.cx(alice,bob)
repetition5.cx(alice,eve)
repetition5.cx(alice,charlie)
repetition5.cx(alice,dave)
repetition5.barrier()

repetition5.cx(bob,alice)
repetition5.cx(bob,eve)
repetition5.cx(bob,charlie)
repetition5.cx(bob,dave)
repetition5.mcx([alice, eve, charlie], aq6[1])
repetition5.mcx([alice, eve, dave], aq6[1])
repetition5.mcx([alice, charlie, dave], aq6[1])
repetition5.mcx([dave, eve, charlie], aq6[1])
repetition5.mcx([alice, eve, charlie,dave], aq6[1])
repetition5.cx(bob,alice)
repetition5.cx(bob,eve)
repetition5.cx(bob,charlie)
repetition5.cx(bob,dave)
repetition5.barrier()

repetition5.cx(eve,alice)
repetition5.cx(eve,bob)
repetition5.cx(eve,charlie)
repetition5.cx(eve,dave)
repetition5.mcx([alice, bob, charlie], aq6[2])
repetition5.mcx([alice, bob, dave], aq6[2])
repetition5.mcx([alice, dave, charlie], aq6[2])
repetition5.mcx([dave, bob, charlie], aq6[2])
repetition5.mcx([alice, bob, charlie, dave], aq6[2])
repetition5.cx(eve,alice)
repetition5.cx(eve,bob)
repetition5.cx(eve,charlie)
repetition5.cx(eve,dave)
repetition5.barrier()

repetition5.cx(charlie,alice)
repetition5.cx(charlie,bob)
repetition5.cx(charlie,eve)
repetition5.cx(charlie,dave)
repetition5.mcx([alice, bob, eve], aq6[3])
repetition5.mcx([alice, bob, dave], aq6[3])
repetition5.mcx([alice, dave, eve], aq6[3])
repetition5.mcx([dave, bob, eve], aq6[3])
repetition5.mcx([alice, bob, eve, dave], aq6[3])
repetition5.cx(charlie,alice)
repetition5.cx(charlie,bob)
repetition5.cx(charlie,eve)
repetition5.cx(charlie,dave)
repetition5.barrier()

repetition5.cx(dave,alice)
repetition5.cx(dave,bob)
repetition5.cx(dave,eve)
repetition5.cx(dave,charlie)
repetition5.mcx([alice, bob, eve], aq6[4])
repetition5.mcx([alice, bob, charlie], aq6[4])
repetition5.mcx([alice, charlie, eve], aq6[4])
repetition5.mcx([charlie, bob, eve], aq6[4])
repetition5.mcx([alice, bob, eve, charlie], aq6[4])
repetition5.cx(dave,alice)
repetition5.cx(dave,bob)
repetition5.cx(dave,eve)
repetition5.cx(dave,charlie)
repetition5.barrier()

repetition5.cx(aq6[0],alice)
repetition5.cx(aq6[1],bob)
repetition5.cx(aq6[2],eve)
repetition5.cx(aq6[3],charlie)
repetition5.cx(aq6[4],dave)
repetition5.cx(alice,bob)
repetition5.cx(alice,eve)
repetition5.cx(alice,charlie)
repetition5.cx(alice,dave)
repetition5.h(alice)
repetition5.barrier()

repetition5.measure([alice[0], bob[0], eve[0],charlie[0],dave[0]],cr)
repetition5.draw(output='mpl')
```
